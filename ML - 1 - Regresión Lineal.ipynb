```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Variable independiente
X = np.array([[65],[68],[70],[70],[72],[75],[78],[80],[82],[85],[88],[90],[95]])
```

```python
# Precio en miles de dólares
y = np.array([62000,66000,68000,72000,74000,76000,76500,78000,84000,86000,91000,92000,98000])
```

```python
plt.rcParams['axes.facecolor'] = 'lightgrey'
plt.figure(figsize=(10,5))
plt.scatter(X,y,color='g', s=100, label='Data')
plt.title('Regresión Lineal')
plt.xlabel("Superficie (m2)")
plt.ylabel("Precio (dólares)")
plt.grid()
```

```python
# Crear el modelo de REGRESIÓN LINEAL
modelo = LinearRegression()

# Entrenar el modelo
modelo.fit(X, y)
```

```python
# Predicción para un inmueble de 76 m2
X_nuevo_1 = np.array([[76]])
y_pred_1 = modelo.predict(X_nuevo_1)

print("Predicción de valor:", y_pred_1)
```

```python
# Predicción para un inmueble de 86 m2
X_nuevo_2 = np.array([[86]])
y_pred_2 = modelo.predict(X_nuevo_2)

print("Predicción de valor:", y_pred_2)
```

```python
plt.figure(figsize=(10,5))
plt.scatter(X,y,color='g', s=100, label='Datos originales')

plt.plot(X_nuevo_1, y_pred_1, color='r', marker='v', label='Predicción 1', markersize=15)
plt.plot(X_nuevo_2, y_pred_2, color='b', marker='v', label='Predicción 2', markersize=15)

plt.title('Regresión Lineal')
plt.legend()
plt.xlabel("Superficie (m2)")
plt.ylabel("Precio (dólares)")
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.scatter(X,y,color='g', s=100, label='Data')

plt.plot(X_nuevo_1, y_pred_1, color='r', marker='v', label='Predicción 1', markersize=15)
plt.plot(X_nuevo_2, y_pred_2, color='b', marker='v', label='Predicción 2', markersize=15)
plt.title('Regresión Lineal')
plt.legend()
plt.xlabel("Superficie (m2)")
plt.ylabel("Precio (dólares)")
plt.grid()

X_line = np.array([[65], [95]])  
y_line = modelo.predict(X_line)  
plt.plot(X_line, y_line, color='black', linestyle=':',label='Linear Regression')
plt.show()
```
